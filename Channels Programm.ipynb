Программа для парсинга значения XML ссылки со списокм всех ТВ-каналов.

```python
def xml_reader(url): 
    # URL вашей XML-ссылки
    # Загрузка данных из URL
    response = requests.get(url)
    print(response.status_code)
    print()
    root = ET.fromstring(response.content)

    data = []

    for channel in root.findall('channel'):
            channel_data = {
                'channel_id':channel.get('id'),
                'epgsrvc_id':channel.get('epgsrvc_id'),
                'display_name':channel.find('display-name').text,
                'week':channel.find('week').text,
                'update':channel.find('update').text,
                'link':channel.find('href').text
            }
            data.append(channel_data)
    # Создание DataFrame из списка словарей
    df = pd.DataFrame(data)
    return df
```

```python
def parse_programme(programme):
    # Создаем словарь для хранения данных о программе
    programme_data = {
        "start": programme.get("start"),
        "stop": programme.get("stop"),
        "channel": programme.get("channel"),
        "id": programme.get("id")
    }
    
    # Проходим по всем тегам внутри <programme>
    for child in programme:
        tag_name = child.tag
        
        # Если тэг может быть повторен несколько раз, создаем список
        if tag_name not in programme_data:
            programme_data[tag_name] = []
        
        # Если тэг уже есть, добавляем к списку
        if isinstance(programme_data[tag_name], list):
            programme_data[tag_name].append(child.text)
        else:
            programme_data[tag_name] = [programme_data[tag_name], child.text]
    
    return programme_data

def main(url):
    tree = fetch_and_parse_xml(url)
    root = tree.getroot()
    
    # Создаем список для хранения всех программ
    programmes = []
    
    for programme in root.findall('programme'):
        programme_data = parse_programme(programme)
        programmes.append(programme_data)
    
    # Преобразуем список словарей в DataFrame
    df = pd.DataFrame(programmes)

    return df

# Пример URL
url = 'http://xmldata.epgservice.ru/EPGService/hs/xmldata/3d6da182-cedb-4315-a685-30294949632/file/000001419?week=2024-08-12'

# Выполняем основную функцию
df = main(url)

# Выводим полученный DataFrame
df
```
